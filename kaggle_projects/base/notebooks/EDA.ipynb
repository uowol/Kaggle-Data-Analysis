```python
# AB TEST
# Assumptions:¶
    # Check normality
    # If Normal Distribution, check homogeneity
# Steps:
    # Split & Define Control Group & Test Group
    # Apply Shapiro Test for normality
    # If parametric apply Levene Test for homogeneity of variances
    # If Parametric + homogeneity of variances apply T-Test
    # If Parametric - homogeneity of variances apply Welch Test
    # If Non-parametric apply Mann Whitney U Test directly


def AB_Test(df, group, target):
    
    # Packages
    from scipy.stats import shapiro
    import scipy.stats as stats
    
    # Split A/B
    groupA = df[df[group] == "A"][target]
    groupB = df[df[group] == "B"][target]

    # Assumption Nomality, 정규성 검정
    # print(shapiro(groupA)[1])           # 샤피로 윌크 검정은 샘플이 5000개 이하일 때 사용, 따라서 이번 데이터에는 부적합
    # print(stats.normaltest(groupA)[1])
    ntA = stats.normaltest(groupA)[1] < 0.05
    ntB = stats.normaltest(groupB)[1] < 0.05
    # H0: Distribution is normal - False
    # H1: Distribution is not normal - True

    if (ntA == False) and (ntB == False):   # 두 그룹 모두 정규성을 만족할 때
        # Parametric Test, 모수적 검정(대부분 정규성 가정으로부터 출발함)
        # Assumption Homogeneity of Variance, 등분산성 검정
        levene_test = stats.levene(groupA, groupB)[1]
        # H0: Homogeneity of Variance - False
        # H1: Heterogeneity of Variance - True
        
        if levene_test >= 0.05:
            # Homogeneity of Variance
            t_test = stats.ttest_ind(groupA, groupB, equal_var=True)[1]
            # H0: The means of the two groups are equal - False
            # H1: The means of the two groups are not equal - True
        else:
            # Heterogeneity of Variance
            t_test = stats.ttest_ind(groupA, groupB, equal_var=False)[1]
            # H0: The means of the two groups are equal - False
            # H1: The means of the two groups are not equal - True
    else:
        # Non-Parametric Test, 비모수적 검정
        t_test = stats.mannwhitneyu(groupA, groupB, alternative='greater')[1]
        # H0: The means of the two groups are equal - False
        # H1: The means of the two groups are not equal - True

    # Result
    temp = pd.DataFrame({
        "AB Hypothesis": [t_test < 0.05],
        "p-value": [t_test],
    })
    temp['Test Type'] = np.where((ntA == False) and (ntB == False), "Parametric Test", "Non-Parametric Test")
    temp["AB Hypothesis"] = np.where(temp["AB Hypothesis"] == True, "Reject H0", "Fail to Reject H0")
    temp["Comment"] = np.where(temp["AB Hypothesis"] == "Reject H0", "B 실험 결과가 A 실험보다 더 크게 나타난다.", "A 실험과 B 실험 결과에 차이가 없다.")

    # Columns
    if (ntA == False) and (ntB == False):
        temp["Homogeneity"] = np.where(levene_test >= 0.05, "Homogeneity of Variance", "Heterogeneity of Variance")
        temp = temp[["Test Type", "Homogeneity","AB Hypothesis", "p-value", "Comment"]]
    else:
        temp = temp[["Test Type", "AB Hypothesis", "p-value", "Comment"]]

    # Print Hypothesis
    print("# A/B Testing Hypothesis")
    print("H0: A == B")
    print("H1: A != B", "\n")
    
    return temp
```

```python
# Apply A/B Testing

# sum_gamerounds ~ version
data = _data.copy()
data["version"] = np.where(data.version == "gate_30", "A", "B")
AB_Test(df=data, group = "version", target = "sum_gamerounds")

# retention_1 ~ version
data = _data.copy()
data["version"] = np.where(data.version == "gate_30", "A", "B")
data[""]
AB_Test(df=data, group = "version", target = "retention_1")
```
